## Funciones: funciones como retorno de otra funci贸n

En Python, las funciones no se limitan a ser solo bloques de c贸digo que realizan tareas espec铆ficas; tambi茅n pueden usarse como datos. Esto significa que puedes definir funciones dentro de otras funciones e incluso devolver funciones desde otras funciones.

Las funciones que devuelven otras funciones se conocen como funciones de orden superior. Estas funciones de orden superior te permiten crear funciones especializadas al vuelo, haciendo tu c贸digo m谩s din谩mico y adaptable. As铆 es como funciona:

1. Una funci贸n puede definirse dentro de otra funci贸n.
2. La funci贸n interna puede capturar y recordar las variables de su funci贸n contenedora (externa). Este concepto se conoce como "cierre".
3. La funci贸n externa puede devolver la funci贸n interna como resultado.

Esta funcionalidad te permite personalizar el comportamiento de las funciones basado en ciertas condiciones o configuraciones, lo cual puede ser extremadamente 煤til en varios escenarios de programaci贸n.

Exploraremos este concepto m谩s a fondo con algunos ejemplos para ver c贸mo las funciones pueden devolverse desde otras funciones.

````python
def greater_function (a, b):
    if a > b:
        return a
    elif b > a:
        return b
````

````python
greater_function (3, 10)
````

````python
def lesser_function (x, y):
    if x > y:
        return y
    elif y > x:
        return x
````

````python
lesser_function (3, 10)
````

````python
def comparison (type_of_comparison): # lesser, greater
    if type_of_comparison == "greater":
        return greater_function
    
    elif type_of_comparison == "lesser":
        return lesser_function
````

````python
comparison("greater")(3, 10)
````

````python
comparison("lesser")(3, 10)
````

6. Ahora que tenemos una referencia a `greater_function`, podemos invocarla con argumentos espec铆ficos, como `greater_function(3, 10)`.

7. La llamada `comparison("greater")(3, 10)` es equivalente a `greater_function(3, 10)`, donde `greater_function` se ejecuta con los argumentos `3` y `10`.

8. Como resultado, devuelve el mayor de los dos n煤meros, que es `10`.

Esto demuestra el concepto de devolver funciones como objetos de primera clase en Python, permitiendo la selecci贸n y ejecuci贸n din谩mica de funciones basadas en condiciones. Usando lambda, dentro de una funci贸n:
